## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import shap
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error
from xgboost import XGBRegressor
```

## 2. Load and Merge Dataset

```python
train = pd.read_csv('train.csv', parse_dates=['Date'])
store = pd.read_csv('store.csv')

df = train.merge(store, on='Store', how='left')
df = df[df['Open'] == 1]
df.sort_values(['Store', 'Date'], inplace=True)
df.head()
```

## 3. Feature Engineering

```python
def create_features(df):
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Week'] = df['Date'].dt.isocalendar().week.astype(int)
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    df['IsWeekend'] = (df['DayOfWeek'] >= 5).astype(int)

    df['Lag_7'] = df.groupby('Store')['Sales'].shift(7)
    df['Lag_14'] = df.groupby('Store')['Sales'].shift(14)
    df['Lag_28'] = df.groupby('Store')['Sales'].shift(28)

    df['RollingMean_7'] = df.groupby('Store')['Sales'].shift(1).rolling(7).mean()
    df['RollingStd_7'] = df.groupby('Store')['Sales'].shift(1).rolling(7).std()
    return df

df = create_features(df)
df.dropna(inplace=True)
df['Sales_Log'] = np.log1p(df['Sales'])
```

## 4. Time-Based Split

```python
split_date = '2015-01-01'
train_df = df[df['Date'] < split_date]
valid_df = df[df['Date'] >= split_date]

features = [
    'Store','DayOfWeek','Promo','StateHoliday','SchoolHoliday',
    'Year','Month','Week','IsWeekend',
    'Lag_7','Lag_14','Lag_28','RollingMean_7','RollingStd_7'
]

X_train = train_df[features]
y_train = train_df['Sales_Log']
X_valid = valid_df[features]
y_valid = valid_df['Sales_Log']
```

## 5. Train Forecast & Quantile Models

```python
def train_model(alpha=None):
    if alpha is None:
        model = XGBRegressor(n_estimators=400, max_depth=8, learning_rate=0.05)
    else:
        model = XGBRegressor(n_estimators=400, max_depth=8,
                             learning_rate=0.05,
                             objective='reg:quantileerror',
                             quantile_alpha=alpha)
    model.fit(X_train, y_train)
    return model

model_point = train_model()
model_q10 = train_model(alpha=0.1)
model_q90 = train_model(alpha=0.9)
```

## 6. Prediction & Uncertainty

```python
pred_log = model_point.predict(X_valid)
pred_q10 = model_q10.predict(X_valid)
pred_q90 = model_q90.predict(X_valid)

valid_df['Prediction'] = np.expm1(pred_log)
valid_df['Lower'] = np.expm1(pred_q10)
valid_df['Upper'] = np.expm1(pred_q90)
valid_df['Actual'] = np.expm1(y_valid)

valid_df['Uncertainty'] = valid_df['Upper'] - valid_df['Lower']
```

## 7. Risk-Based HALT Decision

```python
uncertainty_threshold = valid_df['Uncertainty'].quantile(0.75)

valid_df['HALT'] = (valid_df['Uncertainty'] > uncertainty_threshold).astype(int)
valid_df['Decision'] = valid_df['HALT'].map({0:'Auto-Execute',1:'Human Review'})

valid_df[['Prediction','Actual','Decision']].head()
```

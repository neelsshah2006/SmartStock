```python
import pandas as pd
import numpy as np
import xgboost as xgb
import lightgbm as lgb

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings("ignore")
```

# 1. LOAD DATA

```python
train = pd.read_csv('./data/train.csv')
store = pd.read_csv('./data/store.csv')
```

```python
data = train.merge(store, on='Store', how='left')
data['StateHoliday'] = data['StateHoliday'].astype(str)
```

```python
data = data.drop(columns=['Promo2', 'Promo2SinceWeek', "Promo2SinceYear", "PromoInterval", 'CompetitionOpenSinceMonth'])
data.head()
```

# 2. BASIC CLEANING

```python
data = data[data['Open'] == 1]
data = data[data['Sales'] > 0]

data['Date'] = pd.to_datetime(data['Date'])

data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['WeekOfYear'] = data['Date'].dt.isocalendar().week.astype(int)

data['CompetitionDistance'] = data['CompetitionDistance'].fillna(
    data['CompetitionDistance'].median()
)

data = data.fillna(0)
```

# 3. ENCODE CATEGORICAL FEATURES

```python
for col in ['StoreType','Assortment','StateHoliday']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
```

# 4. TIME SERIES FEATURES

```python
data = data.sort_values(['Store','Date'])
```

```python
data['Average_Sales_7'] = (
    data.groupby('Store')['Sales']
    .transform(lambda x: x.rolling(7).mean())
)

data['Demand_Std_7'] = (
    data.groupby('Store')['Sales']
    .transform(lambda x: x.rolling(7).std())
)

data['Average_Sales_30'] = (
    data.groupby('Store')['Sales']
    .transform(lambda x: x.rolling(30).mean())
)

data['Demand_Std_30'] = (
    data.groupby('Store')['Sales']
    .transform(lambda x: x.rolling(30).std())
)
```

```python
data = data.fillna(0)
```

```python
data.head()
```

# 5. BASE MODEL — XGBOOST FORECAST

```python
features = [
    'Store','DayOfWeek','Promo','SchoolHoliday',
    'StateHoliday','Year','Month','Day','WeekOfYear',
    'CompetitionDistance','StoreType','Assortment',
    'Average_Sales_7', 'Average_Sales_30', 'Demand_Std_7', 'Demand_Std_30'
]
```

```python
X = data[features]
y = data['Sales']
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.4, shuffle=False
)
```

```python
xgb_model = xgb.XGBRegressor(
    n_estimators=500,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred = xgb_model.predict(X_val)

print("Forecast MAPE:", mean_absolute_percentage_error(y_val, y_pred))
```

# 6. INVENTORY SIMULATION

```python
LEAD_TIME = 3
Z_VALUE = 1.65
```

```python
data = data.fillna(0)
```

```python
data['Safety_Stock'] = Z_VALUE * data['Demand_Std_7'] * np.sqrt(LEAD_TIME)
data['Reorder_Point'] = (data['Average_Sales_7'] * LEAD_TIME) + data['Safety_Stock']
```

# 9. PROMOTION ABNORMALITY

```python
data = data.sort_values(['Store','Date'])

data['Past_Promo_Sales'] = (
    data['Sales']
    .where(data['Promo'] == 1)
    .groupby(data['Store'])
    .transform(lambda x: x.shift(1).expanding().mean())
)

data['Past_Normal_Sales'] = (
    data['Sales']
    .where(data['Promo'] == 0)
    .groupby(data['Store'])
    .transform(lambda x: x.shift(1).expanding().mean())
)

data['Expected_Uplift'] = (
    (data['Past_Promo_Sales'] - data['Past_Normal_Sales']) /
    (data['Past_Normal_Sales'] + 1)
)

data = data.fillna(0)

data['Current_Uplift'] = (
    (data['Sales'] - data['Average_Sales_7']) /
    (data['Average_Sales_7'] + 1)
)

data['Promotion_Abnormality'] = (
    data['Current_Uplift'] -
    data['Expected_Uplift']
)
```

# 5. RISK FEATURES (FOR HALT TRAINING)

```python
val_df = X_val.copy()
val_df['Actual'] = y_val.values
val_df['Predicted'] = y_pred

val_df['Error'] = val_df['Actual'] - val_df['Predicted']
val_df['APE'] = abs(val_df['Error']) / (val_df['Actual'] + 1)

val_df['Rolling_MAPE'] = (
    val_df.groupby('Store')['APE']
    .transform(lambda x: x.rolling(7).mean())
)

val_df['Error_Std'] = (
    val_df.groupby('Store')['Error']
    .transform(lambda x: x.rolling(7).std())
)

val_df['Forecast_Confidence'] = 1 / (1 + val_df['Error_Std'])

val_df = val_df.fillna(0)
```

```python
val_df["Volatility_Ratio"] = (
    val_df["Demand_Std_7"] / val_df["Average_Sales_7"]
).replace([np.inf, -np.inf], 0).fillna(0)

val_df["Volatility_Change"] = (
    val_df["Demand_Std_7"] / val_df["Demand_Std_30"]
).replace([np.inf, -np.inf], 0).fillna(0)

val_df["Trend_Shift"] = (
    val_df["Average_Sales_7"] - val_df["Average_Sales_30"]
)

val_df["Trend_Ratio"] = (
    val_df["Average_Sales_7"] / val_df["Average_Sales_30"]
).replace([np.inf, -np.inf], 1).fillna(1)

val_df["Relative_Error"] = (
    np.abs(val_df["Error"]) / val_df["Average_Sales_7"]
).replace([np.inf, -np.inf], 0).fillna(0)

val_df["Prediction_Deviation_Ratio"] = (
    (val_df["Predicted"] - val_df["Average_Sales_7"]) / val_df["Average_Sales_7"]
).replace([np.inf, -np.inf], 0).fillna(0)

val_df["Promo_Risk"] = (
    val_df["Promo"] * val_df["Volatility_Ratio"]
)

val_df["Confidence_Error_Gap"] = (
    val_df["Forecast_Confidence"] * val_df["APE"]
)
```

```python
val_df["Days_Since_Last_Promo"] = (
    val_df.groupby("Store")["Promo"]
    .apply(lambda x: (~x.astype(bool)).cumsum() - 
                     (~x.astype(bool)).cumsum().where(x == 1).ffill().fillna(0))
    .reset_index(level=0, drop=True)
)
```

```python
val_df["Is_Holiday"] = (
    (val_df["SchoolHoliday"] == 1) |
    (val_df["StateHoliday"] != 0)
).astype(int)

val_df["Holiday_Proximity"] = (
    val_df.groupby("Store")["Is_Holiday"]
    .transform(lambda x: x.rolling(7, min_periods=1).max())
)
```

```python
val_df
```

```python
val_df.to_csv("val.csv")
```

```python
risk_features = [
    "Rolling_MAPE",
    "Relative_Error",
    "Forecast_Confidence",
    "Promo_Risk",
    "Volatility_Ratio",
    "Prediction_Deviation_Ratio"
]

scaler = MinMaxScaler()
val_df_scaled = val_df.copy()
val_df_scaled[risk_features] = scaler.fit_transform(val_df[risk_features])

val_df_scaled["Temp_Risk_Score"] = (
    0.25 * val_df["Rolling_MAPE"] +
    0.20 * val_df["Relative_Error"] +
    0.20 * (1 - val_df["Forecast_Confidence"]) +
    0.15 * val_df["Promo_Risk"] +
    0.10 * val_df["Volatility_Ratio"] +
    0.10 * abs(val_df["Prediction_Deviation_Ratio"])
)

threshold = val_df_scaled["Temp_Risk_Score"].quantile(0.70)

val_df["HALT_Label"] = np.where(
    val_df_scaled["Temp_Risk_Score"] > threshold,
    1,
    0
)

print(val_df["HALT_Label"].value_counts(normalize=True))
```

```python
X = val_df[risk_features]
y = val_df["HALT_Label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42, stratify=y
)
```

```python
scale_pos_weight = (y == 0).sum() / (y == 1).sum()

model = lgb.LGBMClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=-1,
    num_leaves=31,
    scale_pos_weight=scale_pos_weight,
    random_state=42
)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import lightgbm as lgb

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('./data/train.csv')
store = pd.read_csv('./data/store.csv')
```

```python
data = train.merge(store, on='Store', how='left')
data['StateHoliday'] = data['StateHoliday'].astype(str)
```

```python
data = data.drop(columns=['Promo2', 'Promo2SinceWeek', "Promo2SinceYear", "PromoInterval", 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear'])
data.head()
```

```python
data = data[data['Open'] == 1]
data = data[data['Sales'] > 0]

data['Date'] = pd.to_datetime(data['Date'])

data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['WeekOfYear'] = data['Date'].dt.isocalendar().week.astype(int)

data['CompetitionDistance'] = data['CompetitionDistance'].fillna(
    data['CompetitionDistance'].median()
)

data = data.fillna(0)
```

```python
for col in ['StoreType','Assortment','StateHoliday']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
```

```python
data = data.sort_values(['Store','Date'])
```

```python
data['Average_Sales_7'] = (
    data.groupby('Store')['Sales']
    .transform(lambda x: x.rolling(7).mean())
)

data['Demand_Std_7'] = (
    data.groupby('Store')['Sales']
    .transform(lambda x: x.rolling(7).std())
)

data['Average_Sales_30'] = (
    data.groupby('Store')['Sales']
    .transform(lambda x: x.rolling(30).mean())
)

data['Demand_Std_30'] = (
    data.groupby('Store')['Sales']
    .transform(lambda x: x.rolling(30).std())
)
```

```python
data = data.fillna(0)
```

```python
data.head()
```

```python
features = [
    'Store','DayOfWeek','Promo','SchoolHoliday',
    'StateHoliday','Year','Month','Day','WeekOfYear',
    'CompetitionDistance','StoreType','Assortment',
    'Average_Sales_7', 'Average_Sales_30', 'Demand_Std_7', 'Demand_Std_30'
]
```

```python
X = data[features]
y = data['Sales']
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.4, shuffle=False
)
```

```python
xgb_model = xgb.XGBRegressor(
    n_estimators=500,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred = xgb_model.predict(X_val)

print("Forecast MAPE:", mean_absolute_percentage_error(y_val, y_pred))
```

```python
val_df = data.copy()[-337736:]
val_df = val_df.drop(columns=['Sales'])
val_df
```

```python
val_df['Actual'] = y_val.values
val_df['Predicted'] = np.ceil(y_pred / 10) * 10

val_df['Error'] = val_df['Actual'] - val_df['Predicted']
val_df['APE'] = abs(val_df['Error']) / (val_df['Actual'] + 1)

val_df['Rolling_MAPE'] = (
    val_df.groupby('Store')['APE']
    .transform(lambda x: x.rolling(7).mean())
)

val_df['Error_Std'] = (
    val_df.groupby('Store')['Error']
    .transform(lambda x: x.rolling(7).std())
)

val_df['Forecast_Confidence'] = 1 / (1 + val_df['Error_Std'])

val_df = val_df.fillna(0)
```

```python
val_df['Past_Promo_Sales'] = (
    val_df['Actual']
    .where(val_df['Promo'] == 1)
    .groupby(val_df['Store'])
    .transform(lambda x: x.shift(1).expanding().mean())
)

val_df['Past_Normal_Sales'] = (
    val_df['Actual']
    .where(val_df['Promo'] == 0)
    .groupby(val_df['Store'])
    .transform(lambda x: x.shift(1).expanding().mean())
)

val_df['Expected_Uplift'] = (
    (val_df['Past_Promo_Sales'] - val_df['Past_Normal_Sales']) /
    (val_df['Past_Normal_Sales'] + 1)
)

val_df['Current_Uplift'] = (
    (val_df['Actual'] - val_df['Average_Sales_7']) /
    (val_df['Average_Sales_7'] + 1)
)

val_df['Promotion_Abnormality'] = (
    val_df['Current_Uplift'] -
    val_df['Expected_Uplift']
)
```

```python
val_df["Volatility_Ratio"] = (
    val_df["Demand_Std_7"] / val_df["Average_Sales_7"]
).replace([np.inf, -np.inf], 0).fillna(0)

val_df["Volatility_Change"] = (
    val_df["Demand_Std_7"] / val_df["Demand_Std_30"]
).replace([np.inf, -np.inf], 0).fillna(0)

val_df["Trend_Shift"] = (
    val_df["Average_Sales_7"] - val_df["Average_Sales_30"]
)

val_df["Relative_Error"] = (
    np.abs(val_df["Error"]) / val_df["Average_Sales_7"]
).replace([np.inf, -np.inf], 0).fillna(0)

val_df["Prediction_Deviation_Ratio"] = (
    (val_df["Predicted"] - val_df["Average_Sales_7"]) / val_df["Average_Sales_7"]
).replace([np.inf, -np.inf], 0).fillna(0)

val_df["Promo_Risk"] = (
    val_df["Promo"] * val_df["Volatility_Ratio"]
)

val_df["Confidence_Error_Gap"] = (
    val_df["Forecast_Confidence"] * val_df["APE"]
)
```

```python
val_df["Days_Since_Last_Promo"] = (
    val_df.groupby("Store")["Promo"]
    .apply(lambda x: (~x.astype(bool)).cumsum() - 
                     (~x.astype(bool)).cumsum().where(x == 1).ffill().fillna(0))
    .reset_index(level=0, drop=True)
)
```

```python
val_df["Demand_Shock_Score"] = (
    val_df["Actual"] - val_df["Average_Sales_30"]
) / (val_df["Demand_Std_30"] + 1)

val_df["Demand_Shock"] = (
    val_df["Demand_Shock_Score"] > 2
).astype(int)
```

```python
val_df["Demand_Momentum"] = (
    val_df["Average_Sales_7"] - val_df["Average_Sales_30"]
) / (val_df["Average_Sales_30"] + 1)

val_df["Demand_Acceleration"] = (
    val_df.groupby("Store")["Actual"]
    .diff()
    .diff()
)

val_df["Demand_Spike"] = (
    val_df["Actual"] >
    val_df["Average_Sales_30"] + 2 * val_df["Demand_Std_30"]
).astype(int)

val_df["Demand_Drop"] = (
    val_df["Actual"] <
    val_df["Average_Sales_30"] - 2 * val_df["Demand_Std_30"]
).astype(int)
```

```python
val_df["Forecast_Uncertainty"] = (
    val_df["Demand_Std_7"] /
    (val_df["Predicted"] + 1)
)
```

```python
val_df["Is_Holiday"] = (
    (val_df["SchoolHoliday"] == 1) |
    (val_df["StateHoliday"] != 0)
).astype(int)

val_df["Holiday_Proximity"] = (
    val_df.groupby("Store")["Is_Holiday"]
    .transform(lambda x: x.rolling(7, min_periods=1).max())
)
```

```python
val_df["Relative_Error"] = np.abs(val_df["Error"]) / (val_df["Actual"] + 1)
```

```python
val_df["Days_Until_Next_Promo"] = (
    val_df.groupby(["Store"])["Promo"]
    .transform(lambda x: x[::-1].cumsum()[::-1])
)
```

```python
val_df["Days_To_Month_End"] = (
    val_df["Date"].dt.daysinmonth - val_df["Date"].dt.day
)
```

```python
val_df["Sales_per_Customer"] = (
    val_df["Actual"] / (val_df["Customers"] + 1)
)

val_df["Customer_Growth"] = (
    val_df.groupby("Store")["Customers"]
    .pct_change()
)

val_df["Customer_Volatility_7"] = (
    val_df.groupby("Store")["Customers"]
    .rolling(7)
    .std()
    .reset_index(level=0, drop=True)
)
```

```python
val_df["Traffic_Spike"] = (
    val_df["Customers"] >
    val_df["Customers"].rolling(30).mean() * 1.5
).astype(int)
```

```python
val_df["Week_sin"] = np.sin(2 * np.pi * val_df["WeekOfYear"] / 52)
val_df["Week_cos"] = np.cos(2 * np.pi * val_df["WeekOfYear"] / 52)

val_df["Day_sin"] = np.sin(2 * np.pi * val_df["DayOfWeek"] / 7)
val_df["Day_cos"] = np.cos(2 * np.pi * val_df["DayOfWeek"] / 7)
```

```python
val_df["Promo_Demand_Risk"] = val_df["Promo"] * val_df["Volatility_Ratio"]
val_df["Forecast_Risk"] = (1 - val_df["Forecast_Confidence"]) * val_df["Volatility_Ratio"]
```

```python
val_df.columns
```

```python
len(val_df.columns)
```

```python
val_df = val_df.fillna(0)
```

```python
forecast_risk = (
    0.5 * np.clip(val_df["Relative_Error"], 0, 1) +
    0.3 * np.clip(val_df["Rolling_MAPE"], 0, 1) +
    0.2 * np.clip(val_df["Prediction_Deviation_Ratio"] / 3, 0, 1)
)

demand_risk = (
    0.4 * np.clip(val_df["Demand_Shock_Score"] / 3, 0, 1) +
    0.3 * np.clip(val_df["Volatility_Ratio"] / 2, 0, 1) +
    0.3 * val_df["Demand_Spike"]
)

promo_risk = (
    0.5 * np.clip(val_df["Promotion_Abnormality"] / 2, 0, 1) +
    0.5 * np.clip(val_df["Promo_Risk"], 0, 1)
)

confidence_risk = (
    0.6 * (1 - val_df["Forecast_Confidence"]) +
    0.4 * np.clip(val_df["Forecast_Uncertainty"], 0, 1)
)

holiday_risk = (
    0.7 * val_df["Is_Holiday"] +
    0.3 * (val_df["Holiday_Proximity"] < 3).astype(int)
)
```

```python
risk_score = (
    0.45 * forecast_risk +
    0.25 * demand_risk +
    0.15 * promo_risk +
    0.10 * confidence_risk +
    0.05 * holiday_risk
)

risk_score = (risk_score - risk_score.min()) / (risk_score.max() - risk_score.min())
```

```python
rng = np.random.default_rng(42)
noise = rng.normal(0, 0.05, len(val_df))
risk_score = risk_score + noise
```

```python
val_df["HALT_Label"] = (risk_score > 0.25).astype(int)
```

```python
val_df['HALT_Label'].value_counts()
```

```python
val_df.columns
```

```python
halt_features = [
    "Forecast_Confidence",
    "Forecast_Uncertainty",
    "Promotion_Abnormality",
    "Promo_Risk",
    "Past_Promo_Sales",
    "Past_Normal_Sales",
    "Days_Since_Last_Promo",
    "Volatility_Ratio",
    "Volatility_Change",
    "Trend_Shift",
    'Demand_Shock_Score',
    'Demand_Spike', 
    'Demand_Drop',
    "Demand_Momentum",
    "Promo_Demand_Risk",
    "Forecast_Risk"
]
```

```python
X_halt = val_df[halt_features]
y_halt = val_df['HALT_Label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_halt, y_halt, test_size=0.4, random_state=42, stratify=y_halt
)
```

```python
scale_pos_weight = (len(y_train) - sum(y_train)) / sum(y_train)
```

```python
halt_model = xgb.XGBClassifier(
    n_estimators=1200,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight,
    eval_metric="auc",
    random_state=42
)
```

```python
halt_model.fit(X_train, y_train)
```

```python
y_prob = halt_model.predict_proba(X_test)[:,1]
```

```python
y_pred = (y_prob > 0.45).astype(int)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```
